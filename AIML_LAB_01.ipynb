2. Using the generated dataset given above. Generate the following graphs and justify the

6. Provide Graph Title, labels for X, Y axis with proper justification and explanation of

```python
people=['kiran','arun','vijay','varun']
age=[25, 30, 35, 40, 45]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]

```

```python
import matplotlib.pyplot as mplot
```

```python
mplot.scatter(weight, height)
mplot.title("Relationship between Weight and Height of Patients")
mplot.xlabel("Height of patients in Centimeters")
mplot.ylabel("Weight of Patients in Kilograms")
mplot.show()
```

```python
mplot.figure(figsize=(12,4))
mplot.title("People weight in kgs",fontsize=16,
          fontstyle='italic')
mplot.bar(x=people,height=weight, width=0.6,
        color='blue',edgecolor='k',alpha=0.6)

mplot.xlabel("persons",fontsize=18)
mplot.xticks(fontsize=16,rotation=35)
mplot.yticks(fontsize=16)
mplot.ylabel("people in health care weight in kgs)",fontsize=15)
mplot.show()
```

```python
import numpy as np
mplot.figure(figsize=(6,6))
mplot.hist(weight,color='green',edgecolor='k', alpha=0.75,bins=5)
mplot.title("Histogram of patient weight",fontsize=20)
mplot.xlabel("patients weight in kgs",fontsize=20)
mplot.xticks(fontsize=10)
mplot.yticks(fontsize=10)
mplot.show()
```
